```python
#3 normal distributions sample 3 times, compute p value
import numpy as np
import sys
from scipy.stats import norm,ttest_1samp
import matplotlib.pyplot as plt
#crappy API, mean=loc, stddev=scale. 
norm1=norm(loc=20,scale=9)
plt.hist(norm1.rvs(size=1000))

norm2=norm(loc=65,scale=9)
plt.hist(norm2.rvs(size=1000),color="red")

#randomly sample 3 from norm1, norm2 calculate mean then difference then pvalue of the difference

from scipy.stats import ttest_ind
num_pvalue=0
for x in range(0,10000):
    #t = ttest_ind(norm1.rvs(size=3),norm2.rvs(size=3),equal_var=False)
    t = ttest_ind(norm1.rvs(size=3),norm1.rvs(size=3),equal_var=False)
    #print(t[1])
    if t[1]<=0.05:
        num_pvalue+=1
print("num_pvalue < 0.05:",num_pvalue) #383 vs same distribution, larger value close to 10k when different distrubution

#this is what he mean
#s by 500 per distribution
n=0
print("1:",norm1.rvs(size=1))
print("2:",norm1.rvs(size=2))
for x in range(0,10000):
    #cant use size=1?
    t = ttest_1samp(norm1.rvs(size=2),popmean=20)
    #print(t[1])
    if t[1]<=0.05:
        n+=1
print(n) #
```
