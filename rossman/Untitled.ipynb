```python
import pandas as pd
df = pd.read_csv("rossman_data/train.csv")
store = pd.read_csv("rossman_data/store.csv")
df.head()
```

```python
df['SalesDate'] = pd.to_datetime(df['Date'])
df.head()
```

```python



store1 = df.loc[ (df["Store"]==1) & (df["Sales"]!=0) ]
print(store1.shape)
store1.head()
```

```python
store1_week1 = store1.loc[(store1["SalesDate"]<="2013-01-13") & (store1["SalesDate"]>"2013-01-06")]
print(store1_week1.shape)
print(store1_week1["Sales"].sum())
store1_week1
```

```python
store1_salesbyweek = store1.groupby([pd.Grouper(key='SalesDate',freq="W")])['Sales'].sum()

#error check the 32952 in the code above
#store to file and do polar plot? 
print(len(store1_salesbyweek))
store1_salesbyweek
```
