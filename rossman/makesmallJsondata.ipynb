```python

import csv
import json
data={}
with open("sales_eachstore_one_week_small.csv", encoding='utf-8') as f:
    csvReader = csv.DictReader(f)
    for rows in csvReader:
        key = rows['Store']
        data[key] = rows

print("data",data)
with open("sales_eachstore_one_week_small.json", 'w', encoding='utf-8') as jsonf:
    jsonf.write(json.dumps(data, indent=4))
```
