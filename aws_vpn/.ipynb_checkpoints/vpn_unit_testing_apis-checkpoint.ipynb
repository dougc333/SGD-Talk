```python
import boto3

ec2 = boto3.client('ec2')

# Retrieves all regions/endpoints that work with EC2
response = ec2.describe_regions()
print('Regions:', response['Regions'])

# Retrieves availability zones only for region of the ec2 object
response = ec2.describe_availability_zones()
print('Availability Zones:', response['AvailabilityZones'])
#opt-in-not-required means this region is enabled by default and you cant shut it off
#this allows for mistakes
#
```

```python
import boto3

session = boto3.session.Session()
print("session, this shows default region:",session)
print(session.profile_name)
print(session.region_name)
print("-------")
part = session.get_available_partitions()
res = session.get_available_resources()
serv = session.get_available_services()
print("partitions:",part)
print("resources:",res)
print("services",serv)
```

```python
import boto3

client = boto3.client('ec2')
sub = client.describe_subnets()
tags = client.describe_tags()
vpcs = client.describe_vpcs()
vpn_gateways = client.describe_vpn_gateways()
#tricky have to parse JSON, no standards
print("subnets:",sub)
print("num subnets:",len(sub['Subnets']))
print("tags:",tags)
print("vpcs:",vpcs)
print("num vpcs:",len(vpcs['Vpcs']))
delete_vpcs=[]
for x in vpcs['Vpcs']:
    print("x:",x)
    print("VpcID:",x['VpcId'])
    delete_vpcs.append(x['VpcId'])
#how to delete a vpc? only wiht name which we assign from tag? Has to be another way
#there is a vpc id when we list resources
print("vpn_gateways:",vpn_gateways)
#print(client.describe_tags())
```

```python
import boto3,inspect,sys
client = boto3.client('ec2')
what_am_i=inspect.getmembers(client)
type(what_am_i)
for x in what_am_i:
    print(type(x),len(x))
    
    
```

```python
import boto3,inspect,sys
client = boto3.client('ec2')
what_am_i=inspect.getmembers(client)
type(what_am_i)
for x in what_am_i:
    if len(x)!=2:
        raise ValueError
    
```

```python
import boto3,inspect,sys
client = boto3.client('ec2')
what_am_i=inspect.getmembers(client)
type(what_am_i)
for x in what_am_i:
    assert(len(x)==2)
```
