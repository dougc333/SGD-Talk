```python
import os,sys,subprocess

class FindImports:
    def __init__(self):
        self.start_dir="/Users/dougchang/Downloads/aws-cli-develop"
    def find_files(self,start_dir):
        files_dirs = os.listdir(start_dir)
        for x in files_dirs:
            #print("x:",x,os.path.join(start_dir,x))
            #print("type:", os.path.join(start_dir,x))
            if(os.path.isdir(os.path.join(start_dir,x))):
                #print("new dir:",os.path.join(start_dir,x))
                self.find_files(os.path.join(start_dir,x))
            else:
                if x[-2:]=="py":
                    #print("file:",x,os.path.join(start_dir,x))
                    here = os.path.join(start_dir,x)
                    cmd = "cat {} | grep import".format(here)
                    #print("cmd:",cmd)
                    ps = subprocess.Popen(cmd,shell=True,stdout=subprocess.PIPE,stderr=subprocess.STDOUT)
                    output = ps.communicate()[0]
                    process_me=output.split(b'\n')
                    for x in process_me:
                        #print("x:",x)
                        if b"import" in x:
                            #print(" stuff at end:",x[x.index(b'import')+6:])
                            if b"boto3" in x[x.index(b'import')+6:]:
                                print("boto3:",x)
                            if b"botocore" in x[x.index(b'import')+6:]:
                                print("botocore:",x)
                                
                            if b"cdk" in x[x.index(b'import')+6:]:
                                print("cdk:",x)
                            if b"aws" in x[x.index(b'import')+6:]:
                                print("aws:",x)
                            
                            
            #    sys.exit()
fi = FindImports()
fi.find_files(fi.start_dir)
```
